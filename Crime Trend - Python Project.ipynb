# Crime Trend Analysis

### Importing Packages

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
from scipy.interpolate import griddata
import seaborn as sns
```

### Function for reading and manipulating the data file

```python
#we can edit UnivName to the desired university name and OnOff to On_Campus or Noncampus when callng this function
def readfile(filename, column_name1, UnivName, column_name2, OnOff):
    df = pd.read_csv(filename, index_col=0) 
    subset_df1 = df.loc[df[column_name1] == UnivName]
    subset_df = subset_df1.loc[subset_df1[column_name2] == OnOff]
    column_names = ['Institution name', ' Campus Name', 'Institution Size','Location','Unitid','Campus ID']
    new_df = subset_df.drop(column_names, axis=1)
    return new_df
```

### Function for plotting all types of crimes in a single graph (trended over time)

```python
#plot a single graph for all crimes in a university
def plotsinglegraph(df, xlbl, ylbl):
    plt.style.use('seaborn-darkgrid')
    #single line plot
    df.plot(x='Survey year', figsize=(30,30), fontsize = 30)
    plt.xlabel(xlbl,fontsize = 40)
    plt.ylabel(ylbl, fontsize = 40)
    plt.legend(fontsize = 20)
```

### Function for creating sub plots of different kinds of crimes (trended over time)

#### Each subplot shows the comparitive data of all kinds of crimes (in grey lines)

```python
#plot subplots of different crimes
def plotgridgraph(df, title):
    plt.style.use('seaborn-darkgrid')
    plt.figure(1, figsize=(30,30))
    # create a color palette
    palette = plt.get_cmap('Set1',16)
    # multiple line plot
    num=0
    for column in df.drop('Survey year', axis=1):
        num+=1
        # Find the right spot on the plot
        plt.subplot(4,4, num)
        # plot every groups, but discreet
        for v in df.drop('Survey year', axis=1):
            plt.plot(df['Survey year'], df[v], marker='', color='grey', linewidth=0.6, alpha=0.3)
        # Plot the lineplot
        plt.plot(df['Survey year'], df[column], marker='', color=palette(num), linewidth=2.4, alpha=0.9, label=column)
        # Add title
        plt.title(column, loc='left', fontsize=20, fontweight=0, color=palette(num) )
    # general title
    plt.suptitle(title, fontsize=24, fontweight=0, color='black', style='italic', y=1.02)
```

## Carnegie Mellon University

### Off Campus Crimes

```python
df = readfile("full_data_final.csv", "Institution name", "Carnegie Mellon University","Location","Noncampus")
plotsinglegraph(df, 'Year', 'Number of crimes')
```

```python
plotgridgraph(df, 'Crime rate')
```
