# Get University Location Information from API

```python
import pandas as pd
import requests
import json
```

## get university list

```python
import numpy as np
University_Folder=np.genfromtxt('University_Folder.csv',delimiter=',',dtype='str')
University_Folder
```

## Request data from API and save them to seperate json files with the name of the universities
### This code will take a while to run

```python
for university in University_Folder:
    result='%20'.join(university.replace('&','_').split('_'))
    response=requests.get('https://api.data.gov/ed/collegescorecard/v1/schools.json?api_key=evae9uBE9KLJd6IlRnqFyQ9rqJpYQxxAh39RYfkd&_fields=id,school.name,school.zip,school.state,school.city,location.lat,location.lon&school.name='+result)
    jdata=response.json()
    f = open(university+'.json', 'w')
    json.dump(jdata,f)
```

# Merge all the Json file to one csv file

## create multiple dataframe from json files

```python
location_dict={}
for university in University_Folder[:-1]:
    d=json.load(open(university+'.json'))
    location_dict[university]=pd.DataFrame.from_dict(d['results'], orient='columns')
```

# merge the dataframes to one dataframe

```python
data=pd.DataFrame()
for university in University_Folder[:-1]:
    data=data.append(location_dict[university],sort=False)
        
data
```

# Save the dataframe to csv

```python
data.to_csv('university_location.csv',sep=',')
```
