#### Creating a DataFrame from the read file

### Reading the full_data_final.csv file to get the University and Crimes information

```python
## Reading in the University-crimes data from a csv and storing it in a Dataframe
crime_data_full=pd.read_csv('full_data_final.csv',delimiter=',')
```

### Merging the two datasets i.e. the University-Crime and University-Location data

### Grouping the complete data over University Names and University Locations

```python
### Grouping columns over Year, University and the University location(Lattitude, Longitude, State and City) to aggregate the crime data
data_grp_university_year=crime_data_full.groupby(['Survey year','Institution name','Lattitude','Longitude','State','City']).sum().reset_index()
```

### Creating the Geographic Map of the Crime Data using Plotly

```python
#### Creating Hover text for map
data_grp_university_year['text'] = 'University name: '+data_grp_university_year['Institution name'] + '' +'Institution size: '+ data_grp_university_year['Institution Size'].astype(str) + ' State: ' + data_grp_university_year['State'] + ' City: ' + data_grp_university_year['City']  + '' + data_grp_university_year['Murder/Non-negligent manslaughter'].astype(str)
```

```python
### Adding average crime rate columns for each year at each crime

data_grp_university_year['Average Murder/Non-negligent manslaughter']=data_grp_university_year['Murder/Non-negligent manslaughter']/data_grp_university_year['Institution Size']

data_grp_university_year['Average Negligent manslaughter']=data_grp_university_year['Negligent manslaughter']/data_grp_university_year['Institution Size']

data_grp_university_year['Average Sex offenses - Forcible']=data_grp_university_year['Sex offenses - Forcible']/data_grp_university_year['Institution Size']

data_grp_university_year['Average Rape']=data_grp_university_year['Rape']/data_grp_university_year['Institution Size']

data_grp_university_year['Average Fondling']=data_grp_university_year['Fondling']/data_grp_university_year['Institution Size']

data_grp_university_year['Average Sex offenses - Non-forcible']=data_grp_university_year['Sex offenses - Non-forcible']/data_grp_university_year['Institution Size']

data_grp_university_year['Average Incest']=data_grp_university_year['Incest']/data_grp_university_year['Institution Size']

data_grp_university_year['Average Statutory rape']=data_grp_university_year['Statutory rape']/data_grp_university_year['Institution Size']

data_grp_university_year['Average Robbery']=data_grp_university_year['Robbery']/data_grp_university_year['Institution Size']

data_grp_university_year['Average Aggravated assault']=data_grp_university_year['Aggravated assault']/data_grp_university_year['Institution Size']

data_grp_university_year['Average Burglary']=data_grp_university_year['Burglary']/data_grp_university_year['Institution Size']

data_grp_university_year['Average Motor vehicle theft']=data_grp_university_year['Motor vehicle theft']/data_grp_university_year['Institution Size']

data_grp_university_year['Average Arson']=data_grp_university_year['Arson']/data_grp_university_year['Institution Size']

data_grp_university_year['Average Hate_Crimes']=data_grp_university_year['Hate_Crimes']/data_grp_university_year['Institution Size']





```

```python
### Defining color scale for map

scl = [ [0,"rgb(5, 10, 172)"],[0.35,"rgb(40, 60, 190)"],[0.5,"rgb(70, 100, 245)"],[0.6,"rgb(90, 120, 245)"],[0.7,"rgb(106, 137, 247)"],[1,"rgb(220, 220, 220)"] ]
```

```python
### Creating a dictionary of Crime data for each year(key)
data_grp_university={}
for survey_year in data_grp_university_year['Survey year'].unique():
    data_grp_university[survey_year]=data_grp_university_year.loc[data_grp_university_year['Survey year']==survey_year]
```

```python
### Asking for user input for getting the map corresponding to particular year
which_year=input('Enter a year between 2001 and 2016 to view average crimes as per locations')
```

```python
### Making a list of crime categories
list_categories=['Murder/Non-negligent manslaughter', 'Negligent manslaughter',
       'Sex offenses - Forcible', 'Rape', 'Fondling',
       'Sex offenses - Non-forcible', 'Incest', 'Statutory rape', 'Robbery',
       'Aggravated assault', 'Burglary', 'Motor vehicle theft', 'Arson',
       'Hate_Crimes']
```

```python
### Asking user input of crime category to display the map for
which_category=input('Enter a category from above list to view average crime rates as per universities')
```

```python
### Creating the variable which would be mapped on the graph(the average crime rate(crime count/Insitution size))
which_category='Average '+which_category
```

```python
which_category
```

```python
### would give the Crime Data for various universities for the specified year
data_grp_university[int(which_year)]
```

```python
### Creating the title name for the map
titlename=which_category+' '+which_year
titlename
```

```python
### Creating the parameter to be provided for iplot function of plotly
data = [ dict(
        type = 'scattergeo',
        locationmode = 'USA-states',
        lon = data_grp_university[int(which_year)]['Longitude'],
        lat = data_grp_university[int(which_year)]['Lattitude'],
        text = data_grp_university[int(which_year)]['text'],
        mode = 'markers',
        marker = dict(
            size = 8,
            opacity = 0.8,
            reversescale = True,
            autocolorscale = False,
            symbol = 'triangle',
            line = dict(
                width=1,
                color='rgba(102, 102, 102)'
            ),
            colorscale = scl,
            cmin = 0,
            color = data_grp_university[int(which_year)][which_category],
            cmax = data_grp_university[int(which_year)][which_category].max(),
            colorbar=dict(
                title=titlename
            )
        ))]
```

```python
data
```

```python
heading_map=which_category+' at universities in '+which_year+'(hover for University names)'
heading_map
```

```python
### Creating the parameter to be provided for iplot function of plotly

layout = dict(
        title = heading_map,
    colorbar='True',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showland = True,
            landcolor = "rgb(250, 250, 250)",
            subunitcolor = "rgb(217, 217, 217)",
            countrycolor = "rgb(217, 217, 217)",
            countrywidth = 0.5,
            subunitwidth = 0.5
        ),
    )            
```

```python
layout
```

```python
### Creating the parameter to be provided for iplot function of plotly
fig = dict( data=data, layout=layout )
```

```python
## Plot the data on the map
py.plotly.iplot( fig,validate=False, filename='Map for crime rates' )
```
